#***NOSQL(Not Only SQL) DATABASE:***

NoSQL databases are non-relational. NoSQL is used to store unsructured and semit-structured data. It has horizontal scalabilty. Data can be stored in different formats:

JSON documents (like MongoDB)

Key-value pairs (like Redis)

Graphs (like Neo4j)

Column stores (like Cassandra)

*   ***Unstructured Data:***

 Data that has no clear format or structure. It’s not in rows/columns and is often hard to search or organize.

 ***Example:***
 Emails, Images, Videos, PDFs, Chat messages, Word documents.

 It is used in Social media (photos, videos, posts), Customer support chats
 and Articles and blogs

*   ***Semi-structured Data:***

 Data that’s partially organized not in tables, but still has labels or tags to separate pieces of info.

 ***Example:***
 Web APIs, Logs, Configuration files and NoSQL databases like MongoDB.

*   ***Horizontal Scalability:***

 Add more machines/servers to handle the load. Like adding more workers to a team. Instead of one powerful server, use 4 smaller servers together.

 ### ***Working of NoSQL:***

 NoSQL uses KEY-VALUE store. It is like they have 2 fields one is for key and other contains values like if it is a product, key field will contain the barcode and value field will contain it's name. Then the value field will contains some more data like product price, range, description etc. This further data is stored in the JSON format.
If single database server is not enough to store all the data and queries we will split the workload among 2 or more servers. Apple runs NoSQL database that consists of 75,000 servers. These are called Partitions. If our daabase is split into 1000 of servers then how we will know which part of data is stored in which server? How we will identfy data? Then we use primary key. Key determines on which partition an item will be stored. NoSQL uses Hash function to convert each item's primary key into a number that falls into a fixed range between 0 and 100. The hash value and range is used to determine where to store an item. If you us 1 server the range will be from 0 to 100 but when there are 2 servers the range will become 0-50 for 1 server and 50 to 100 for other.1 server will be responsible for all items with a hash function between 0 to 50 and other will store from 50 to 100. Finding data with the key is quiet efficient but when we have to find it with certain amount it becomes difficult but it is handled well by the SQL database.

#***Web Scrapping:***
Web scraping is the process of automatically extracting data from websites. It involves using software tools or scripts to load a webpage, parse its content (usually HTML), and collect specific pieces of information such as:

Text

Links

Images

Prices

Tables

Product details

News articles, etc.

###***Common Tools/Libraries for Web Scraping (Python):***

1. ***BeautifulSoup*** – for parsing HTML and XML.

2. ***Requests*** – for sending HTTP requests.

3. ***Selenium*** – for interacting with dynamic websites (JavaScript-heavy).

4. ***Scrapy*** – a powerful framework for large-scale scraping.

5. ***Playwright*** – modern tool for browser automation.

###***Steps for Web Scraping (Python):***

1. ***Go to the Website:***

 Just like you type a website URL in your browser, the web scraping code also "visits" a website.

2. ***Get the Page Content:***

 The scraper sees the page's behind-the-scenes code — the HTML (like reading the skeleton of the page).

3. ***Look for the Data you Need:***

 The scraper searches through the HTML to find the exact data you want — for example, prices, names, news headlines, etc.

4. ***Take out the Data:***

 Once it finds the data, it pulls it out (this is called "extracting" the data).

5. ***Save the Data:***

 Finally, the scraper saves this data in a file — like a spreadsheet (CSV), a text file, or a database — so you can use it later.

```python
import requests
import re

url = "https://en.wikipedia.org/wiki/Indeed"
headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
html = response.text

# Remove all script and style tags manually
html = re.sub(r'<(script|style).*?>.*?</\1>', '', html, flags=re.DOTALL)

# Remove all HTML tags
text = re.sub(r'<[^>]+>', '', html)

# Remove extra spaces and blank lines
cleaned_text = "\n".join([line.strip() for line in text.splitlines() if line.strip()])

print(cleaned_text[:2000])  # Preview
```

#2. ***Web Scrapping Using BeautifulSoup (Picking What you Want from the Page):***

It helps you find and extract the useful stuff (like product name, price).

Think of it like:

##Libraries:

requests: Lets us download the webpage.

urljoin: Helps build a complete image link, in case it’s broken or partial.

Image, display: Used in Google Colab or Jupyter Notebook to show images in the output.
##Website URL Setup:
url: This is the website you want to read/scrape.

headers: We pretend to be a browser (to avoid getting blocked). This is like saying:

"Hi! I'm Chrome. Can I see the page?"

##What happens when you write a Python script to scrape?:

By default, a Python request like requests.get(url) sends minimal headers, often missing User-Agent.

The server sees: "No User-Agent or weird headers → maybe a bot → block or throttle."

But if you add headers mimicking Chrome (like a real browser’s User-Agent), it looks more human → usually allowed.

##HTML Parser:
The HTML parser is the tool that reads the raw HTML code (all the tags and text) and converts it into a structured format (called a "parse tree" or "DOM tree").

This makes it easy to find elements like h1, p, img, etc., by their tags, classes, or attributes.



```python
from bs4 import BeautifulSoup
import requests
from urllib.parse import urljoin
from IPython.display import Image, display

# --- Setup ---
url = "https://en.wikipedia.org/wiki/Indeed"
headers = {
    "User-Agent": "Mozilla/5.0"
}

# --- Fetch and parse the page ---
response = requests.get(url, headers=headers)

#This line cleans up the raw HTML and turns it into something Python can understand and work with easily.Go get the web page, then clean it up so I can search through it and pick out what I need.
soup = BeautifulSoup(response.text, 'html.parser')
#This line extracts all the visible text from the web page, removing HTML tags and stores it in the variable all_text. Strip remove any extra spaces at the start and end of each text piece.
all_text = soup.get_text(separator='\n', strip=True)

# --- Extract and display images ---
images = []
print("🖼️ Displaying Images:\n")
for img_tag in soup.find_all('img'):
    src = img_tag.get('src')
    if src:
        full_img_url = urljoin(url, src)
        images.append(full_img_url)
        try:
            display(Image(url=full_img_url))
        except:
            pass  # skip broken or inaccessible image URLs

# --- Print preview of text ---
print("\n✅ Text Preview:\n")
print(all_text[:1000])  # Only preview first 1000 characters
```

#Printing NULL for Missing Values:
country_name_tag.text gets the text inside the HTML tag, but it might have extra spaces or newlines.

.strip() removes these spaces/newlines.



```python
def get_details(country_one): #One country’s details, but easily repeated for many
  country_name = country_one.find('h3', class_='country-name').text.strip()
  capital_name = country_one.find('span', class_='country-capital').text
  population = country_one.find('span', class_='country-population').text
  area = country_one.find('span', class_='country-area').text
  return country_name,capital_name, population, area
  # Use .text.strip() if the tag exists, else "null"
  country_name = country_name_tag.text.strip() if country_name_tag and country_name_tag.text.strip() else "null"
  capital_name = capital_name_tag.text.strip() if capital_name_tag and capital_name_tag.text.strip() else "null"
  population = population_tag.text.strip() if population_tag and population_tag.text.strip() else "null"
  area = area_tag.text.strip() if area_tag and area_tag.text.strip() else "null"

  return country_name, capital_name, population, area
```

```python
def get_soup(url):
    """Takes URL and returns a soup object"""
    resp = requests.get(url)
    if resp.status_code == 200: #The number 200 here is an HTTP status code that means "OK"
        return bs(resp.text)
    else: return None


def get_countries(url):
    """Extact details from all the countries"""
    soup = get_soup(url)
    country_all = soup.find_all('div', class_="col-md-4 country")
    countries = []
    for country in country_all:
        countries.append(get_details(country))

    return countries
```

```python
url = 'https://www.scrapethissite.com/pages/simple/'
countries = get_countries(url)
len(countries) #250 countries
```

#***SQLite Database:***
SQLite is a serverless, self-contained, file-based database.

That means it doesn’t require any database server to run (unlike MySQL or PostgreSQL).

It stores all the data in a single .db file on your computer.

Think of it like a spreadsheet saved as a file, but one that you can access using Python and SQL queries.

```python
import sqlite3
```

# ***Creating and Connecting Database and Tables:***

Firstly we are creating a database named countries and then we are connecting it with the sqlite. This creates or opens a SQLite database file called countries.db. ***conn*** is a variable that holds an object specifically, a connection object from the sqlite3 library.

This ***cursor()*** creates a cursor object from the connection.

A cursor is like a pointer or a pen that you use to write/read SQL commands inside the database.

All SQL operations (like CREATE, INSERT, SELECT) are done via the cursor.

Until you run commit(), changes are temporary (you can undo them by rollback()).

After commit(), the changes are permanently written to the countries.db file.

```python
# Connect to (or create) SQLite database
conn = sqlite3.connect('countries.db')
cursor = conn.cursor()

# Create table
cursor.execute('''CREATE TABLE countries (id INTEGER PRIMARY KEY AUTOINCREMENT,name TEXT,capital TEXT,population TEXT,area TEXT)''')
conn.commit()
```

#***Inserting Data:***

```python
for country in country_all:
    name, capital, population, area = get_details(country)
    cursor.execute('''
        INSERT INTO countries (name, capital, population, area) VALUES (?, ?, ?, ?) ''', (name, capital, population, area))

conn.commit() #commit the data into the database
```

#***Displaying Stored Data:***
***SELECT * means:*** "Get all columns".
***FROM countries*** means: "from the countries table"

So it’s saying: "Give me all the data from the countries table".


***cursor.fetchall()***
This grabs all the results from the query we just executed.

It returns a list of rows, where each row is a tuple of values from the database and then displays the result using print function.

```python
cursor.execute('SELECT * FROM countries') #This doesn’t return data directly, it just tells the database what to do.
for row in cursor.fetchall():
    print(row)
```
